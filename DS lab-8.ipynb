```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
iris=pd.read_csv('IRIS.csv')
iris
```

```python
iris.head()
```

```python
df1=iris.loc[iris['Species'] =='Iris-setosa']
df1
```

```python
t_count=np.count_nonzero(df1.SepalLengthCm)
t_count
```

```python
temp=df1.SepalLengthCm.unique()
temp
```

```python
freq=[]
prob=[]
count=0

for i in temp:
    for j in df1['SepalLengthCm']:
        if i==j:
            count=count+1
    freq.append(count)
    prob.append(count/t_count)
    count=0


print(prob)
print(temp)
print(freq)
```

```python
df3=pd.DataFrame(
    {'sepal length': temp,
     'frequency':freq,
     'probability':prob})
df3
```

```python
expect_value=0
i=0
for value in temp:
    expect_value= expect_value + (value * prob[i])
    i=i+1

print('expected value=',expect_value)
```
