```python
import numpy as np
import pandas as pd
```

```python
a = [0,0,1,0,1] #Sum = 2
b = [1,0,0,1,1] #Sum = 3
a1 = np.array(a)
a2 = np.array(b)
d1_1 = [a1,a2]
d1_2 = pd.DataFrame(d1_1,index = ['Array-1','Array-2'])
d1_2
```

```python
s1 = np.sum(a1)
s2 = np.sum(a2)
d2_1 = [[s1,s2],[s1*3,s2*3]]
d2_2 = pd.DataFrame(d2_1,columns = ['Array-1','Array-2'],index = ['Sum', 'Mul'])
d2_2
```

```python
a3 = np.array(b)
bool_val = (a1 == a2) #If Element Of Array-1 Equals To Array-2
d3_1 = [a1,a2,np.logical_and(a1,a3),np.logical_or(a1,a3),np.logical_not(a1,a3),bool_val]
d3_2 = pd.DataFrame(d3_1,index = ['Array-1','Array-2','Logical AND','Logical OR','Logical NOT','Boolean Operation'])
d3_2
```

```python
td1 = [[1,0,1],[0,1,0],[0,1,1]]
td2 = [[1,1,1],[1,1,0],[0,0,1]]
t = np.array(td1)
d = np.array(td2)
td = np.array(td2)
bool_val0 = (t[0] == d[0])
bool_val1 = (t[1] == d[1])
bool_val2 = (t[2] == d[2])
d3_1 = [[t[0],d[0],np.logical_and(t[0],td[0]),np.logical_or(t[0],td[0]),np.logical_not(t[0],td[0]),bool_val0],[t[1],d[2],np.logical_and(t[1],td[1]),np.logical_or(t[1],td[1]),np.logical_not(t[1],td[1])],[t[2],d[2],np.logical_and(t[2],td[2]),np.logical_or(t[2],td[2]),np.logical_not(t[2],td[2])]]
d3_2 = pd.DataFrame(d3_1,columns = ['Array-1','Array-2','Logical AND','Logical OR','Logical NOT','Boolean Operation'])
d3_2
```

```python
d4_1 = [[t[0],d[0],'[0:2]',t[0,0:2],d[0,0:2]],[t[1],d[1],'[0:1]',t[1,0:1],d[1,0:1]]]
d4_2 = pd.DataFrame(d4_1,columns = ['Array-1','Array-2','Slice Range','Sliced Array-1','Sliced Array-2'])
d4_2
```
