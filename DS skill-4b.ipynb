```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
pd.options.display.float_format='{:.5f}'.format
pd.options.display.max_columns=None
pd.options.display.max_rows= None
np.random.seed(100)
```

```
nm=pd.read_csv("/content/NHANES_demographic.csv")
nm.head()
```

```
#dist plot
for i in nm.select_dtypes(include='float'):
  sns.distplot(nm[i])
  plt.show()
```

```
#line plot
#line plot
plt.plot(nm['RIDRETH1'], nm['RIDRETH3'])
```

```
#area plot for single column
ax = nm.plot.area(y='RIDRETH3')
```

```
#bar plot
plt.figure(figsize=(12,8))
sns.barplot(x=nm['RIDRETH1'],y= nm['RIDRETH3'])
plt.show()
```

```
#area plot
ax = nm.plot.area()
```

```
#scatter plot
sca = plt.scatter(nm['RIDRETH1'], nm['RIDRETH3'])
```

```
#histogram plot
plt.hist(x=nm['RIDRETH1'])
```
