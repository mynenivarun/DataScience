```python
from scipy import stats
##assigning the value to sub-datframe from main dataframe having with their product line values 
fa = df.loc[df['Product line']=='Fashion accessories']
hab = df.loc[df['Product line']=='Health and beauty']
ea = df.loc[df['Product line']=='Electronic accessories']
hal = df.loc[df['Product line']=='Home and lifestyle']
fab = df.loc[df['Product line']=='Food and beverages']
st = df.loc[df['Product line']=='Sports and travel']

#assigning modes to the variables 
fa_mode = stats.mode(fa['Product line'])
hab_mode = stats.mode(hab['Product line'])
ea_mode = stats.mode(ea['Product line'])
hal_mode = stats.mode(hal['Product line'])
fab_mode = stats.mode(fab['Product line'])
st_mode = stats.mode(st['Product line'])

#printing the modes of product lines
print("mode of Fashion accessories :",fa_mode,end='\n\n')
print("mode of Health and beauty :",hab_mode,end='\n\n')
print("mode of Electronic accessories :",ea_mode,end='\n\n')
print("mode of Home and lifestyle :",hal_mode,end='\n\n')
print("mode of Food and beverages :",fab_mode,end='\n\n')
print("mode of Sports and travel :",st_mode,end='\n\n')
```
