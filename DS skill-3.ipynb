```python
import pandas as pd
import numpy as np
hospital_ratings=pd.read_csv('Hospital General Information.csv',encoding="ISO-8859-1")
hospital_ratings
```

```python
#1 cleaning data 
hospital_ratings.isnull()
```

```python
hospital_ratings.isnull().sum()
hospital_ratings.notnull()
```

```python
#2 Hospital Analysis
dff=hospital_ratings.groupby(['Hospital Ownership','Hospital Type']).size()
dff
```

```python
#3 Total number of Hospitals in each 'State'
df2=hospital_ratings.groupby(['State']).size()
df2
```

```python
#4 Country wise hospital numbers in "TX"
df3=hospital_ratings[hospital_ratings['State'] == 'TX']
df3
```

```python
#5 Distribution of 'Hospital Type' and 'Ownership in Texas'
distri=hospital_ratings[['State','Hospital Ownership']]
distri1=distri[hospital_ratings['State']=='TX']
distri1
```

```python
#6.	Find Timeliness of care national comparison footnote and its hospital
time=pd.crosstab(hospital_ratings['Hospital Name'],hospital_ratings['Timeliness of care national comparison footnote']=='Above the national average')
time
```

```python
#7.	Hospital rating
ratings=hospital_ratings[['Hospital Name','Hospital overall rating']]
ratings
```

```python
#8.	Find Acute Care Hospitals with rating (4) along with  state
Acute=hospital_ratings.loc[(hospital_ratings["Hospital Type"]=="Acute Care Hospitals")&(hospital_ratings["Hospital overall rating"]=="4"),["State"]]
Acute
```

```python
#9.	Which type of hospitals has highest Non-availability of Safety of Care data?
non_available_data = hospital_ratings.groupby('Hospital Type')['Safety of care national comparison'].count()
non_available_data.sort_values(ascending=False)
```

```python
#10 the different types of Patient experience national comparison and list their Provider ID?
dfnew=hospital_ratings.groupby('Safety of care national comparison')[['Provider ID']].count()
dfnew
```
